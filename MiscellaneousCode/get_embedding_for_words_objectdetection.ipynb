```python

import json
from openai import OpenAI
import os
import pickle
import random
import json

load_dotenv("variables.env")
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
with open("/home/liuxiao/TuringGithub/XiaoData/ZeroShotJudge/groundtruths_d_objdet.json",'r') as f:
    groundtruths_d_objdet = json.load(f)
```

```python
word_openai_embedding_dict = {}
for k,v in groundtruths_d_objdet.items():
    for entry in v:
        key = (entry['imageID'],entry['machine_groundtruth'])
        word_openai_embedding_dict[key] = [entry['response']]
```

```python
word_openai_embedding_dict
```

```python
for key, word in word_openai_embedding_dict.items():
    response = client.embeddings.create(
            model="text-embedding-3-large",
            input=[word[0]]
        )
    word_openai_embedding_dict[key].append(response.data[0].embedding) 
#     break
```

```python
with open("/home/liuxiao/TuringGithub/XiaoData/objectdetection_openai_embedding_dict.pkl",'wb') as f:
    pickle.dump(word_openai_embedding_dict,f)
```
