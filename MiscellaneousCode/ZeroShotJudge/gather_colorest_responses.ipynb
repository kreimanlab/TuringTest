```python
import os

def get_gt_from_pagename(pagename):
    # example:  'page_name': '568_e_0.html/',
    return {'a':'Human','b':'Machine','c':'Machine','d':'Machine','e':'Catch'}[pagename.split('_')[1]]

def get_model_from_pagename(pagename):
    # example:  'page_name': '568_e_0.html/',
    return {'a':'', 'b':'azure','c':'histogram','d':'google'}[pagename.split('_')[1]]

response_path = '/home/liuxiao/TuringGithubRunnable/dominant_color_recognition/Feedback_Color/static/responses_color'

groundtruth_d = {}
for html in os.listdir(response_path):
    with open(response_path+'/'+html,'r') as f:
        text = f.readlines()[0]
    imgid, code, _ = html.split('_')
    if code == 'e': continue
    gt = get_gt_from_pagename(html)
    model = get_model_from_pagename(html)
    response = text.split('Speaker Response: ')[-1].split('</h4>')[0]
    
    imgpath = '{}.jpg'.format(imgid)
    d = {'imageID':imgpath, 'response':response, 'groundtruth':gt.lower(), 'human_groundtruth':'', 'machine_groundtruth':model}
    groundtruth_d.setdefault(imgpath,[]).append(d)

print(groundtruth_d)    
```

```python
import random

for k, trials in groundtruth_d.items():
    random.shuffle(trials)
```

```python
import json

with open('/home/liuxiao/TuringGithub/XiaoData/ZeroShotJudge/groundtruths_d_colorest.json','w') as f:
    json.dump(groundtruth_d,f)
```
