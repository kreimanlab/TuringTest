```python
import json
import random
```

```python
path = "/home/liuxiao/TuringGithub/imagecaption/MturkExps/image_captioning_task/Turing_Test/static/data/main_caption_data.json"
with open(path) as f:
    d = json.load(f)
```

```python
d
```

```python
gt_format = {'response_speaker': 'machine','groundtruth': 'human', 'machine_groundtruth': ''}
captions_d = {}
groundtruths_d = {}
for fn,v in d.items():
    groundtruths = []
    captions = []
    for cap in v['human']:
        groundtruths.append({'imageID':fn,'cap': cap[0],'groundtruth': 'human', 'human_groundtruth': cap[1],'machine_groundtruth': ''})
    for model, cap in v['machine'].items():
        groundtruths.append({'imageID':fn,'cap': cap,'groundtruth': 'machine', 'human_groundtruth': '','machine_groundtruth': model})
    random.shuffle(groundtruths)    
#     print(groundtruths)
    for i,gt in enumerate(groundtruths):
        captions.append('{}. {}\n'.format(i+1,gt['cap']))
    captions_d[fn] = ''.join(captions)
    groundtruths_d[fn] = groundtruths
#     break

with open('/home/liuxiao/TuringGithub/XiaoData/ZeroShotJudge/captions_d.json','w') as f:
    json.dump(captions_d,f)

with open('/home/liuxiao/TuringGithub/XiaoData/ZeroShotJudge/groundtruths_d.json','w') as f:
    json.dump(groundtruths_d,f)
```
