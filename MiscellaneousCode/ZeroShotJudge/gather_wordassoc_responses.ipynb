```python
import os

def get_gt_from_pagename(pagename):
    # example:  'page_name': '568_e_0.html/',
    return {'a':'Human','b':'Machine','c':'Machine','d':'Machine','e':'Machine','f':'Machine','g':'Catch'}[pagename.split('_')[1]]

def get_model_from_pagename(pagename):
    # example:  'page_name': '568_e_0.html/',
    return {'a':'','b':'word2vec','c':'gpt2','d':'gpt3emb','e':'gpt3promptcurie','f':'gpt3promptdav'}[pagename.split('_')[1]]

response_path = '/home/liuxiao/TuringGithubRunnable/wordAssociation/Feedback_Word/static/responses_wordassoc'

groundtruth_d = {}
for html in os.listdir(response_path):
    with open(response_path+'/'+html,'r') as f:
        text = f.readlines()[0]
    imgid, code, _ = html.split('_')
    if code == 'g': continue
    gt = get_gt_from_pagename(html)
    model = get_model_from_pagename(html)
    response = text.split('Association: ')[-1].split('</p>')[0].strip().lower()
    
    cueWord = text.split('; Association: ')[0].split('Cue: ')[-1].strip().lower()
    wordID = imgid
    d = {'cueWord':cueWord, 'response':response, 'groundtruth':gt.lower(), 'human_groundtruth':'', 'machine_groundtruth':model}
    groundtruth_d.setdefault(wordID,[]).append(d)

print(groundtruth_d)    
```

```python
import random

for k, trials in groundtruth_d.items():
    random.shuffle(trials)
```

```python
import json

with open('/home/liuxiao/TuringGithub/XiaoData/ZeroShotJudge/groundtruths_d_wordassoc.json','w') as f:
    json.dump(groundtruth_d,f)
```
