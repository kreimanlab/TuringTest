## Object Detection

### compile ChatGPT responses

```python
import os
import pickle
```

```python
imgd = {}
for response in os.listdir("/home/liuxiao/TuringGithub/dominant_color_recognition/MturkExps/dominant_color_recognition/static/responses"):
    imgd[response.split('_')[0]] = imgd.get(response.split('_')[0],0)+1
print(len(imgd.keys()))
```

```python
with open("/home/liuxiao/TuringGithub/XiaoData/all_results/chatgpt_object_detection.pkl",'rb') as f:
    chatgpt_obj = pickle.load(f)
```

```python
exceptions = ['548.jpg','548']
chatgpt_obj_parsed = {}
root = '/home/liuxiao/TuringGithub/wordAssociation/MturkExp-20240730T132439Z-001/MturkExp/word_associations/MturkExp/NO_GENDER/Word_ChatGPT/static/responses_wordassoc'
for k, v in chatgpt_obj.items():
    if k in exceptions:
        continue
    objects = [_.split('.')[-1].strip().lower() for _ in v.split('\n')]
    fn = k.split('.')[0]
    chatgpt_obj_parsed[fn] = ', '.join(objects)
#     print(objects)
    
for imgid in imgd.keys():
    if imgid in exceptions:
        continue
    fn = "{}_f_0.html".format(imgid)

    content = '''<center> <h4>Question: What do you see in this image? <br> Speaker Response: {}</h4> </center>\n'''.format(chatgpt_obj_parsed[imgid])
    with open(root+fn,'w') as f:
        f.writelines(content)
    print(fn,content)
```

```python
parameter_list = {}
for fn in os.listdir(root):
    model = fn.split('_')[1]
    parameter_list.setdefault(model,[]).append(fn)
    
for k, l in parameter_list.items():
    print('var {}List = {}\n'.format(k,l))
```

```python
exceptions = []#['548.jpg','548']
chatgpt_color_parsed = {}
colors = "red, pink, orange, yellow, purple, green, blue, brown, gray, black, white".split(', ')
root = '/home/liuxiao/TuringGithub/dominant_color_recognition/MturkExps/Color_ChatGPT/static/responses_color_chatgpt/'
for k, v in chatgpt_obj.items():
    if k in exceptions:
        continue
    color = v.replace('.','').lower().strip()
    fn = k.split('.')[0]
    chatgpt_color_parsed[fn] = color
    if color not in colors:
        print(k,v)
#     print(objects)
    
for imgid in imgd.keys():
    if imgid in exceptions:
        continue
    fn = "{}_e_0.html".format(imgid)

    content = '''<center> <h4>Question: What is the dominant color in this image? <br> Speaker Response: {}</h4> </center>\n'''.format(chatgpt_color_parsed[imgid])
    with open(root+fn,'w') as f:
        f.writelines(content)
    print(fn,content)
```
