```python
import csv
import sqlite3
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np

Hgen_responses_all = []
AIgen_responses_all = []
Hpgen_responses_all = []
AIpgen_responses_all = []

with open("./savedResults/compiled.json", 'r') as f:
    CompiledResults = json.load(f)
key = 'inlab_gender'
contents = CompiledResults[key]

# Iterating over keys
for trialid in contents:
        
    gtlabel = contents[trialid]['trialData']['groundtruth']   
    predictlabel = contents[trialid]['trialData']['response_speaker']
    # gender on true label
    conv1_Agen = contents[trialid]['trialData']['response_gender']
    if conv1_Agen == 'female':
        conv1_Agen = 'F'
    elif conv1_Agen == 'male':
        conv1_Agen = 'M'

    if gtlabel == 'human':
        Hgen_responses_all.append(conv1_Agen)
    elif gtlabel == 'machine':
        AIgen_responses_all.append(conv1_Agen)

    #'''
    # gender on predicted label
    if predictlabel == 'human':
        Hpgen_responses_all.append(conv1_Agen)
    elif predictlabel == 'machine':
        AIpgen_responses_all.append(conv1_Agen)

    #'''
    
            

print("==========================================")
print("H (true label) classified as F tot",Hgen_responses_all.count('F'))
print("H (true label) classified as M tot",Hgen_responses_all.count('M'))
ratioHF = Hgen_responses_all.count('F')/(Hgen_responses_all.count('F')+Hgen_responses_all.count('M'))
print("Percentage H (true label) classified as F", np.round(ratioHF,2)*100,'%')

print("AI (true label) classified as F tot",AIgen_responses_all.count('F'))
print("AI (true label) classified as M tot",AIgen_responses_all.count('M'))
ratioAIF =AIgen_responses_all.count('F')/(AIgen_responses_all.count('F')+AIgen_responses_all.count('M'))
print("Percentage AI classified (true label) as F", np.round(ratioAIF,2)*100,'%')

print("H (pred label) classified as F tot",Hpgen_responses_all.count('F'))
print("H (pred label) classified as M tot",Hpgen_responses_all.count('M'))
ratioHpF = Hpgen_responses_all.count('F')/(Hpgen_responses_all.count('F')+Hpgen_responses_all.count('M'))
print("Percentage H (pred label) classified as F", np.round(ratioHpF,2)*100,'%')

print("AI (pred label) classified as F tot",AIpgen_responses_all.count('F'))
print("AI (pred label) classified as M tot",AIpgen_responses_all.count('M'))
ratioAIpF = AIpgen_responses_all.count('F')/(AIpgen_responses_all.count('F')+AIpgen_responses_all.count('M'))
print("Percentage AI (pred label) classified as F", np.round(ratioAIpF,2)*100,'%')

tot_F = (Hgen_responses_all+AIgen_responses_all).count('F')/len(Hgen_responses_all+AIgen_responses_all)
print("Overall percentage classified as F",np.round(tot_F,2)*100,'%')
```
