```python
import csv
import sqlite3
import pandas as pd
import json
```

```python
con = sqlite3.connect('participants.db')

# Load the data into a DataFrame
df = pd.read_sql_query("SELECT * from TuringConversationalAI", con)
```

```python
df
```

```python
num_workers = df.shape[0]
print("we currently have "+str(num_workers)+" workers")
```

```python
for w in range(num_workers):
    print("#### Worker {} ####".format(w))
    # load worker's data
    data = json.loads(df["datastring"][w])
    Wgender = data['questiondata']['Wgender']
    Wage = data['questiondata']['Wage']
    Wlanguage = data['questiondata']['Wlanguage']
    Wdegree = data['questiondata']['Wdegree']
    Wfield = data['questiondata']['Wfield']
    print("W data",Wgender,Wage,Wlanguage,Wdegree,Wfield)
    # load responses to conv1
    conv1_hit = data['data'][3]['trialdata']['hit']
    conv1_len = data['data'][3]['trialdata']['hit_len']
    conv1_Aclass = data['data'][3]['trialdata']['Aclass']
    conv1_Aage = data['data'][3]['trialdata']['Aage']
    conv1_Agender = data['data'][3]['trialdata']['Agender']
    conv1_Bclass = data['data'][3]['trialdata']['Bclass']
    conv1_Bage = data['data'][3]['trialdata']['Bage']
    conv1_Bgender = data['data'][3]['trialdata']['Bgender']
    print("conv1:",conv1_hit,conv1_len,conv1_Aclass,conv1_Aage,conv1_Agender,conv1_Bclass,conv1_Bage,conv1_Bgender)
    # load responses to conv1
    conv2_hit = data['data'][4]['trialdata']['hit']
    conv2_len = data['data'][4]['trialdata']['hit_len']
    conv2_Aclass = data['data'][4]['trialdata']['Aclass']
    conv2_Aage = data['data'][4]['trialdata']['Aage']
    conv2_Agender = data['data'][4]['trialdata']['Agender']
    conv2_Bclass = data['data'][4]['trialdata']['Bclass']
    conv2_Bage = data['data'][4]['trialdata']['Bage']
    conv2_Bgender = data['data'][4]['trialdata']['Bgender']
    print("conv2:",conv2_hit,conv2_len,conv2_Aclass,conv2_Aage,conv2_Agender,conv2_Bclass,conv2_Bage,conv2_Bgender)
    
```
