```python
import csv
import json
import sqlite3

results = {'inlab':[], 'inlab_gender':[], 'inlab_nogender':[], 'mturk':[], 'machine':[], 'combined':[]}


############# for jason file processing
jsonpath = './Data_all/'
filenames = {'mturk':['turing_live_rebuttal_newcap']}

#Images 1-250 are from COCO test set,
# images 251-500 are nocaps val in-domain,
# images 501-750 are nocaps val near-domain,
# images 751-1000 are nocaps val out-domain
```

```python
with open(ROOT+"/Plot/savedResults/compiled_feedback_rebuttal_allresults.json","r") as f:
    content = json.load(f)
results['mturk-old'] = content['mturk']
with open(ROOT+"/Plot/savedResults/compiled_newcap.json","r") as f:
    content = json.load(f)
results['mturk-new'] = content['mturk']
#save combined results
count = 0
trials_dict = {}
for ke in ['mturk-old','mturk-new']:
    for it in results[ke]:
        count = count + 1
        trialin = results[ke][it]
        trials_dict['Trial_'+str(count)] = trialin

results['merged'] = trials_dict

print('All data is extracted.')
for key in results:
    print(key)
    print('There are [' + str(len(results[key])) + ']')

#save the compiled results as jason
with open(ROOT+"/Plot/savedResults/compiled_newcap_merged.json", "w") as fp:
    json.dump(results,fp)

print('Compilation completed! Results are saved to folder!')
```
